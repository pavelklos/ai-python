## Popular Python Data Science Libraries

**Pandas**: A powerful data manipulation library that provides data structures like DataFrames for efficiently working with structured data. It excels at data cleaning, transformation, and analysis with functions for filtering, grouping, merging datasets, and handling missing values.

**Seaborn**: A statistical visualization library built on matplotlib that creates attractive, informative statistical graphics. It offers high-level interfaces for drawing common statistical plots with built-in themes for professional-looking visualizations.

**Bokeh**: An interactive visualization library focused on web-based, interactive plots. It creates JavaScript-powered visualizations that work well in web browsers and notebooks with features like zooming, panning, and hovering for data exploration.

**Key differences between Seaborn and Bokeh**:

- *Purpose*: **Seaborn** focuses on statistical visualization; **Bokeh** specializes in interactive, web-based graphics
- *Output*: **Seaborn** produces static plots; **Bokeh** creates interactive visualizations
- *Backend*: **Seaborn** uses matplotlib; **Bokeh** generates JavaScript visualizations
- *Use case*: **Seaborn** is better for publication-quality statistical charts; **Bokeh** excels at interactive dashboards and web applications

Short explanation of each package and the differences between **Seaborn** and **Bokeh**:

```python
df1 = pd.DataFrame(data_frame, columns = ['Name', 'Wage', 'Value'])

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

df1['Wage'] = df1['Wage'].replace('[\â‚¬,]', '', regex = True).apply(value_to_float)
df1['Value'] = df1['Value'].replace('[\â‚¬,]', '', regex = True).apply(value_to_float)

df1['difference'] = df1['Value'] - df1['Wage']
df1.sort_values('difference', ascending=False)
```

## bokeh (interactive plots and applications in the browser from Python)
https://bokeh.org/<br>
https://docs.bokeh.org/en/latest/<br>
Interactive visualization library for modern web browsers.<br>
It provides elegant, concise construction of versatile graphics and affords high-performance interactivity across large or streaming datasets.<br>
Bokeh can help anyone who wants to create interactive plots, dashboards, and data applications quickly and easily.

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Name", "@Name")]
)

p = figure(title="Soccer 2019", x_axis_label='Wage', y_axis_label='Value',
           # plot_width=700, plot_height=700, tools=[TOOLTIPS])
           width=700, height=700, tools=[TOOLTIPS])
# p.circle('Wage', 'Value', size=10, source=df1)
p.scatter('Wage', 'Value', size=10, source=df1)
show(p)
```
